```python
# find courses that got no enrollment
student_list = np.setdiff1d(courses['course_id'],y['course_id'])
courses[courses['course_id'].isin(student_list)]
```

```python
#find students who did not enroll into any courses
student_list = np.setdiff1d(student['student_id'],y['student_id'])
student[student['student_id'].isin(student_list)]
```

```python
# print student name  -> partner name for all enrolled students
student.merge(student,how='inner',left_on='partner',right_on='student_id')
```

```python
#find top 3 students who did most number enrollments
y.merge(student).groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# find the top 3 students who spend most amount of money on courses
y.merge(student,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```
