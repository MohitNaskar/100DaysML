```python
# find the number of movies that start with a

def foo(group):
    df = group[group['Series_Title'].str.startswith('A')]
    return df.count()
```

```python
genre.apply(foo)
```

```python
#find ranking of each movie in the group according to IMDB score

def rank_movie(group):
    group['IMDB_Rating'].rank(ascending= True)
    return group

genre.apply(rank_movie)
```
